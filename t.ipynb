```python
import matplotlib.pyplot as plt
import numpy as np

# 创建一个简单的散点图
x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 3, 5]

plt.scatter(x, y)

# 在图上的特定点添加注释
plt.annotate('Point A', xy=(2, 4), xytext=(3, 4.5),
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0.5'))

plt.annotate('Point B', xy=(4, 3), xytext=(4.5, 2),
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=-0.5'))

# 显示图
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# 创建一个有向图
G = nx.DiGraph()

# 添加节点
G.add_node("A")
G.add_node("B")
G.add_node("C")

# 添加带标签的边
G.add_edge("A", "B", label="Edge from A to B")
G.add_edge("B", "C", label="Edge from B to C")

# 获取边的标签信息
edge_labels = {(source, target): data['label'] for source, target, data in G.edges(data=True)}

# 绘制图形
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=500, node_color='lightblue')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='red')
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# 创建一个有向图
G = nx.DiGraph()

# 添加节点和带标签的边
G.add_node("0", label="Start", partition=0)
G.add_node("1", label="Decision 1\n(Condition 1)", partition=1)
G.add_node("2", label="Decision 2\n(Condition 2)", partition=1)
G.add_node("Outcome1", label="Outcome 1\n(Result 1)", partition=2)
G.add_node("Outcome2", label="Outcome 1\n(Result 1)", partition=2)

G.add_edge("0", "1", label="True")
G.add_edge("0", "2", label="False")
G.add_edge("1", "Outcome1", label="True")
G.add_edge("2", "Outcome2", label="False")

# 获取边的标签信息
edge_labels = {(source, target): data['label'] for source, target, data in G.edges(data=True)}

# 获取节点的标签信息
node_labels = {node: data['label'] for node, data in G.nodes(data=True)}

# 绘制图形
pos = nx.multipartite_layout(G, subset_key="partition")
nx.draw(G, pos, labels=node_labels, with_labels=True, node_size=800, node_color='lightblue')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='red')
plt.show()
```
